```python
import random
import matplotlib.pyplot as plt
from collections import Counter

def simulate_game():
    position = 1
    steps = []

    for _ in range(24):
        step = random.randint(1, 2)
        position += step
        steps.append(position)

    return steps

def simulate_multiple_games(num_simulations):
    results = []

    for _ in range(num_simulations):
        final_positions = simulate_game()
        results.extend(final_positions)

    return results

def plot_line_graph(results):
    step_counts = Counter(results)
    steps, counts = zip(*sorted(step_counts.items()))

    plt.plot(steps, counts, color='purple', marker='o', linestyle='-', markersize=4)
    plt.xlabel('Step Number, n')
    plt.ylabel('Count')
    plt.title('The count obtained of step numbers across 1000 Simulations of a 24-throw coin game')
    plt.grid(True)
    plt.show()

# simulate 1000 games
num_simulations = 1000
simulation_results = simulate_multiple_games(num_simulations)

# plot the line graph
plot_line_graph(simulation_results)
```
